Correallation matrix of observed behaviours in protypes and the amount of stress per self evalution

```python

import pandas as pd
from scipy import stats

# Sorting
sorting = pd.read_csv('./sorting.csv')
tower = pd.read_csv('./tower.csv')

df = tower
df = df.dropna(subset=['self assesment total'])
df = df.fillna(0)

print(F'Self evalution mean: {df["self assesment total"].mean()}')
print(F'Self evalution SD: {df["self assesment total"].std()}')


#Colums to compare
colums = [
    'total action',
    'Raise eyebrows',
    'Frown',
    'Look away',
    'Biting lip',
    'Lip corner pull',
    'Lip suck',
    'Blow',
    'Fidgeting with pieces',
    'Groom',
    'Scratching',
    'Hand to face',
    'Folding arms',
]

#dict for results
results = {
    'name': [],
    'M': [],
    'SD': [],
    'Correlation coefficient': [],
    'P-value': [],
}

# loop through colums
for colum in colums:

  mean = df[colum].mean()
  std = df[colum].std()
  correlation, p_value = stats.pearsonr(df['self assesment total'], df[colum])

  results['name'].append(colum)
  results['M'].append(mean)
  results['SD'].append(std)
  results['Correlation coefficient'].append(correlation)
  results['P-value'].append(p_value)


resultsDF = pd.DataFrame(results)

resultsDF

```
